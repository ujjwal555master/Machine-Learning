### A Date and time consist many things like date -> date, month, year, weekend, previous similarly with time

# Working with date

```python
#convert it into date datatype
#Working with date
```

```python
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

## Extract year

```python
date['year'] = date['date'].dt.year
date.head()
```

## Extract month

```python
date['month'] = date['date'].dt.month
date.head()
```

```python
date['month_name'] = date['date'].dt.month_name()
date.head()
```

## Extract day

```python
date['day'] = date['date'].dt.day
date.head()
```

```python
#day of week 
date['day_dow'] = date['date'].dt.dayofweek
date.head()
```

```python
date['day_dow_name'] = date['date'].dt.day_name()
date.head()
```

```python
# is weekend?
date['weeken'] = np.where(date['day_dow_name'].isin(['Sunday','Saturday']),1,0)
date.drop(columns=['product_id','city_id','orders'], inplace=True)
date.head()
```

```python
date['week_of_the_year'] = date['date'].dt.isocalendar().week
```

```python
date.head()
```

## Extract quater

```python
date['quarter'] = date['date'].dt.quarter
```

```python
date.head()
```

```python
date['semester'] = np.where(date['quarter'].isin([1,2]),1,2)
```

```python
date.head()
```

# Extract time elasped between two days

```python
import datetime
today = datetime.datetime.today()
today
```

```python
(today-date['date']).dt.days
```

```python
# Months passed
np.round((today-date['date']) / np.timedelta64(1, 'm'),0)
```

# Working with time

```python
time.info()
```

```python
# Converting to datetime datatype
time['date'] = pd.to_datetime(time['date'])
```

```python
time.head()
```

```python
time['hour'] = time['date'].dt.hour
time['min'] = time['date'].dt.minute
time['sec'] = time['date'].dt.second

time.head()
```

```python
#extract time part
time['time'] = time['date'].dt.time
time.head()
```

```python
today - time['date']
```

```python
# in seconds
(today - time['date'])/np.timedelta64(1,'s')
```

```python
# in minutes
(today - time['date'])/np.timedelta64(1,'m')
```

```python
# in hour
(today - time['date'])/np.timedelta64(1,'h')
```
