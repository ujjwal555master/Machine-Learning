### 3. Bi and multi variet analysis

```python
titanic = pd.read_csv('titanic.csv')
iris = sns.load_dataset('iris')
flight = sns.load_dataset('flights')
titanic.head()
```

```python
iris.head()
```

```python
flight.head()
```

# we see scatter ploting (Numerical data - Numerical data)

```python
sns.scatterplot(titanic, x='Age', y='Fare', hue='Pclass', style='Survived')
```

## Bar plot (Numerical - categorical)

```python
sns.barplot(titanic, x='Pclass', y='Age')
```

## Box plot (Numerical - categorical)

```python
sns.boxplot(titanic, x='Pclass', y='Age')
```

```python
sns.boxplot(titanic, x='Pclass', y='Age', hue='Survived')
```

## Distplot (Numerical - Categorical)

```python
sns.displot(titanic[titanic['Survived']==0]['Age'])
```

```python
pd.crosstab(titanic['Pclass'], titanic['Survived'])
```

## Cluster map and heat map(Categorical - Categorical )

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```

```python
sns.clustermap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```

```python
sns.pairplot(iris, hue='species')
```

```python
plt.plot(flight.pivot_table(values='passengers', index='month', columns='year'))
```
